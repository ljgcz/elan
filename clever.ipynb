By considering high-quality and low-quality momentum stocks, we can differentiate between two types of performance:

- **High-quality momentum stocks** show "slow and steady" outperformance over long periods of time.
- **Low-quality momentum stocks** might not show any momentum for an extended period, only to surge upwards later.

This approach significantly reduces the noise generated by short-term movements that are unlikely to be repeated in the future.

To construct this strategy, we will select stocks based on the highest returns over the following periods:

- 1-month price returns
- 3-month price returns
- 6-month price returns
- 1-year price returns
